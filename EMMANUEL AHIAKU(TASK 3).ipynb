## Q9

```python
df7["month"].value_counts()
```

```python
monthly_sales = df7.groupby("month")["price"].sum().reset_index()
monthly_sales
```

```python
monthly_sales.sort_values(by= "price", ascending=True)
```

```python
plt.plot(monthly_sales["month"], monthly_sales["price"], marker="o")
plt.xlabel("month")
plt.ylabel("price(USD)")
plt.title("Total Sales Made In Each Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

Noticeable trend: From the graph above, it can be seen that the month of July has the highest peak indicating much sales made and September has the lowest peak also showing the least sales made.
